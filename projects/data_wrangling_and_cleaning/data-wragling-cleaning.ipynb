```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset contoh
data = {
    'Nama': ['Alice', 'Bob', 'Charlie', 'David', np.nan, 'Eve', 'Frank', 'Grace', 'Henry', 'Ivy'],
    'Usia': [25, 30, np.nan, 35, 40, 29, np.nan, 50, 23, 31],
    'Gaji': [50000, 54000, 62000, np.nan, 72000, 49000, 57000, 80000, np.nan, 52000],
    'Kota': ['Jakarta', 'Bandung', 'Surabaya', 'Jakarta', 'Bandung', 'Surabaya', np.nan, 'Jakarta', 'Bandung', 'Surabaya'],
    'Status': ['Single', 'Married', 'Single', 'Single', np.nan, 'Married', 'Single', 'Married', 'Single', 'Married']
}

df = pd.DataFrame(data)
```

```python
# Menampilkan data awal
print("Data sebelum dibersihkan:")
print(df)

# 1. Menghapus baris dengan nilai yang hilang secara keseluruhan
df_cleaned = df.dropna()

# 2. Mengisi nilai yang hilang pada kolom numerik dengan median
df['Usia'].fillna(df['Usia'].median(), inplace=True)
df['Gaji'].fillna(df['Gaji'].median(), inplace=True)

# 3. Mengisi nilai yang hilang pada kolom kategori dengan modus
df['Nama'].fillna('Tidak Diketahui', inplace=True)
df['Kota'].fillna(df['Kota'].mode()[0], inplace=True)
df['Status'].fillna(df['Status'].mode()[0], inplace=True)

# 4. Menghapus duplikasi jika ada
df.drop_duplicates(inplace=True)

# 5. Visualisasi distribusi usia dan gaji
plt.figure(figsize=(10, 4))
sns.histplot(df['Usia'], bins=5, kde=True, color='blue')
plt.title("Distribusi Usia")
plt.show()

plt.figure(figsize=(10, 4))
sns.histplot(df['Gaji'], bins=5, kde=True, color='green')
plt.title("Distribusi Gaji")
plt.show()
```
